```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
#output the graph in the same shell
```

```python
#Joint Plot

iris=sns.load_dataset('iris')
#print iris
iris.head()
#iris('species').value_counts()
```

```python
#Joint Plot

sns.jointplot(x='sepal_length',y='sepal_width',data=iris,size=10)
```

```python
#Facet Grid Plot
#variable in data to map plot spects to different colors

sns.FacetGrid(iris,hue='species',size=5)\
.map(plt.scatter,'sepal_length','sepal_width')\
.add_legend()
```

```python
#Bar Graph

sns.boxplot(x='species',y='petal_length',data=iris)
```

```python
#sns.boxplot(x='species',y='petal_length',data=iris,jitter=True)
```

```python
#Violin Plot

sns.violinplot(x='species',y='petal_length',data=iris)
```

```python
#Pair Plot

sns.pairplot(data=iris,kind='scatter')
```

```python
#Heat Map

df=pd.DataFrame(np.random.rand(5,5),columns=['a','b','c','d','e'])
print(df)
corr=df.corr()
print(corr)
sns.heatmap(corr)
```

```python
flights=sns.load_dataset('flights')
#print(flights)
flights.head()

flights=flights.pivot('month','year','passengers')
sns.heatmap(flights)
```

```python
sns.heatmap(flights,annot=True,fmt='d',linewidth=.5,cmap='YlGnBu')
```

```python
#Dist Plot

x=np.random.normal(size=100)
#print(x)
#sns.distplot(x)
sns.distplot(x,kde=False,rug=True)
```

```python
sns.distplot(x,bins=5,kde=False,rug=True)
```

```python
#Kernal Density Estimation
#It is auseful tool for plotting the shape of distribution
sns.distplot(x,bins=5,hist=False,rug=True)
```
